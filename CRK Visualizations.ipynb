```python
import warnings
warnings.filterwarnings('ignore')

# Import Neccessary libraries
import numpy as np 
import pandas as pd 

# Import Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the decimal format
pd.options.display.float_format = "{:.2f}".format
```

### <b><span style='color:#16C2D5'>|</span> Input the data</b> 

```python
df = pd.read_csv("CUSTOM__BINARY_diabetes_012_health_indicators_BRFSS2015 copy.csv")
```

```python
df.head()
```

```python

# To get a list of all column headers
column_headers = df.columns.tolist()

print(column_headers)
```

```python
# Remove Unneccessary value [0.00195%]
df = df[df['Sex'] != 'Other']
```

### <b><span style='color:#16C2D5'>|</span> Visualazations</b> 

```python
#Bar plot for gender
# import seaborn as sns
# import matplotlib.pyplot as plt

# Map numeric values to descriptive labels for the purpose of plotting
df['Gender'] = df['Sex'].map({0: 'Male', 1: 'Female'})

# Plot using 'Gender' for both x and hue; this will automatically generate a legend
sns.countplot(x='Gender', hue='Gender', data=df, dodge=False)  # `dodge=False` to avoid separating the bars

# Customize the legend
plt.legend(title='Gender')

plt.title('Gender Distribution')
plt.xlabel('Gender')  # Optionally adjust the label if needed
plt.ylabel('Count')   # Optionally adjust the label if needed
plt.show()
```

```python
#Bar plot for Smoker

# 'Smoker' column where 0 represents nonsmoker and 1 smoker
# Map numeric values to descriptive labels for the purpose of plotting
df['Smoker'] = df['Smoker'].map({0: 'NonSmoker', 1: 'Smoker'})

# Plot using 'Smoker' for both x and hue; this will automatically generate a legend
sns.countplot(x='Smoker', hue='Smoker', data=df, dodge=False)  # `dodge=False` to avoid separating the bars

# Customize the legend
plt.legend(title='Smoker')

plt.title('Smoker Distribution')
plt.xlabel('Smoker')  # Optionally adjust the label if needed
plt.ylabel('Count')   # Optionally adjust the label if needed
plt.show()
```

```python
#Bar plot for GenHlth
df['GenHlth'] = df['GenHlth'].map({0: 'Great', 1: 'Fine', 2: 'Nuetral', 3: 'Not Good', 4: 'Bad', 5: 'Terrible'})

# Plot using 'Gender' for both x and hue; this will automatically generate a legend
sns.countplot(x='GenHlth', hue='GenHlth', data=df, dodge=False)  # `dodge=False` to avoid separating the bars

# Customize the legend
plt.legend(title='GenHlth')

plt.title('General Health')
plt.xlabel('GenHlth')  # Optionally adjust the label if needed
plt.ylabel('Count')   # Optionally adjust the label if needed
plt.show()
```

```python
# Distribution plot for BMI
sns.distplot(df['BMI'], bins=30)
plt.title('BMI Distribution')
plt.show()
```

```python
# Count plots for binary variables
for col in ['HighBP', 'HeartDiseaseorAttack', 'Diabetes_012']:
    sns.countplot(x=col, data=df)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
# Boxplot BMI vs Diabetes classification
sns.boxplot(x='Diabetes_012', y='BMI', data=df)
plt.title('BMI vs Diabetes')
plt.show()
```

```python
# Boxplot Age vs Diabetes classification
sns.boxplot(x='Diabetes_012', y='Age', data=df)
plt.title('Age vs Diabetes')
plt.show()
```

```python
# Interaction between gender, BMI and diabetes
sns.boxplot(x='Diabetes_012', y='BMI', hue='Sex', data=df)
plt.title('BMI Distribution by Diabetes Status and Gender')
plt.show()
```
